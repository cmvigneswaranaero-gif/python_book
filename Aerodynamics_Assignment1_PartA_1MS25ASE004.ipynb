Case 1 Simple For Loop

```python
# -----------------------------------------
# Define constants (single values)
# -----------------------------------------

S   = 47        # Wing area (stored as a number)
CL  = 0.5       # Lift coefficient (constant value)
rho = 1.225     # Air density (single float value)

# -----------------------------------------
# Define a LIST of velocities
# -----------------------------------------
# Square brackets [] create a Python list
# Each element is accessed one at a time in the loop

V = [30, 40, 50, 60, 70, 80, 90, 100,
     120, 140, 160, 180, 200, 220,
     240, 260, 280, 300]

# -----------------------------------------
# for-loop syntax
# -----------------------------------------
# for v in V:
#   → v takes one value from the list V in each iteration
#   → loop runs until all elements in V are used

for v in V:

    # v**2 → square of velocity
    # * → multiplication operator
    # Lift equation implemented as a single expression

    Lift = 0.5 * rho * v**2 * S * CL

    # f-string syntax:
    # f"..." allows variable values inside {}
    # :.2f limits float output to 2 decimal places

    print(f"Speed = {v} m/s --> Lift = {Lift:.2f} N")

```

Case 2: Looping over two lists with zip()


```python
# -----------------------------------------
# Constants (already explained earlier)
# -----------------------------------------

S  = 47          # Wing area
CL = 0.5         # Constant lift coefficient

# -----------------------------------------
# Two LISTS of equal length
# -----------------------------------------

V   = [100, 150, 200, 250, 300, 350]          # Velocity list
rho = [1.225, 1.112, 1.007, 0.909, 0.819, 0.736]  # Density list

# -----------------------------------------
# zip() usage
# -----------------------------------------
# zip(V, rho) pairs elements position-wise:
# (V[0], rho[0]), (V[1], rho[1]), ...
# Loop stops at the shortest list automatically

for v, r in zip(V, rho):

    Lift = 0.5 * r * v**2 * S * CL

    print(f"Speed = {v} m/s, Density = {r} kg/m^3 --> Lift = {Lift:.2f} N")
```

Case 3: Index-based Looping (Lift vs Velocity)

```python
# -----------------------------------------
# Constants (already explained earlier)
# -----------------------------------------

S  = 47          # Wing area
CL = 0.5         # Constant lift coefficient

# -----------------------------------------
# Lists (same length)
# -----------------------------------------

V   = [100, 150, 200, 250, 300, 350]          # Velocity list
rho = [1.225, 1.112, 1.007, 0.909, 0.819, 0.736]  # Density list

# -----------------------------------------
# Index-based looping
# -----------------------------------------
# range(len(V)) generates indices: 0, 1, 2, ...
# i is used to access elements from both lists

for i in range(len(V)):

    v = V[i]       # Access velocity using index i
    r = rho[i]     # Access density using same index

    Lift = 0.5 * r * v**2 * S * CL

    print(f"Speed = {v} m/s, Density = {r} kg/m^3 --> Lift = {Lift:.2f} N")
```

Case 4: Nested Loop (Lift vs Velocity for Multiple Densities)

```python
# -----------------------------------------
# Constants (already defined earlier)
# -----------------------------------------

S  = 47          # Wing area
CL = 0.5         # Constant lift coefficient

# -----------------------------------------
# Lists
# -----------------------------------------

rho = [1.225, 1.112, 1.007, 0.909, 0.819, 0.736]  # Density list
V   = [30, 40, 50, 60, 70, 80, 90, 100,
       120, 140, 160, 180, 200, 220,
       240, 260, 280, 300]                        # Velocity list

# -----------------------------------------
# Nested loop
# -----------------------------------------
# Outer loop: selects one density at a time
# Inner loop: runs through all velocities for that density

for r in rho:

    for v in V:

        Lift = 0.5 * r * v**2 * S * CL

        print(f"Density = {r} kg/m^3   Speed = {v} m/s --> Lift = {Lift:.2f} N")
```

Case 5: while Loop (Lift vs Velocity until Threshold)

```python
# -----------------------------------------
# Constants
# -----------------------------------------

S   = 20       # Wing area
CL  = 0.5      # Lift coefficient
rho = 1.2      # Air density

# -----------------------------------------
# Initial value
# -----------------------------------------
# Starting velocity

V = 30

# -----------------------------------------
# Lift threshold
# -----------------------------------------
# Loop will stop once Lift exceeds this value

threshold = 20000

# -----------------------------------------
# while loop
# -----------------------------------------
# while True:
#   → creates an infinite loop
#   → loop runs until 'break' is executed

while True:

    # Lift calculation
    L = 0.5 * rho * V**2 * S * CL

    print(f"Velocity = {V} m/s --> Lift = {L:.0f} N")

    # if-condition inside while loop
    # checks stopping criteria

    if L > threshold:
        print("Lift exceeded threshold!")
        break     # immediately exits the while loop

    # Update variable
    # V += 5 is shorthand for V = V + 5

    V += 5
```

Case 6: while Loop with Condition in Header

```python
# -----------------------------------------
# Constants
# -----------------------------------------

S   = 20
CL  = 0.5
rho = 1.2

# -----------------------------------------
# Initial velocity
# -----------------------------------------

V = 10

# -----------------------------------------
# Initial Lift calculation
# -----------------------------------------
# Required before entering the while loop

L = 0.5 * rho * V**2 * S * CL

# -----------------------------------------
# while-condition loop
# -----------------------------------------
# Loop continues as long as condition is TRUE

while L <= 20000:

    print(f"Velocity = {V} m/s --> Lift = {L:.0f} N")

    V += 5                     # increment velocity
    L = 0.5 * rho * V**2 * S * CL   # update Lift
```

Case 6: NumPy – Single Density, Multiple Speeds (Lift vs Velocity)

```python
# -----------------------------------------
# Import NumPy
# -----------------------------------------
import numpy as np

# -----------------------------------------
# Constants
# -----------------------------------------

S   = 47          # Wing area
CL  = 0.5         # Constant lift coefficient
rho = 1.225       # Single density (sea level)

# -----------------------------------------
# NumPy array of velocities
# -----------------------------------------
# np.array() converts list into NumPy array
# Enables element-wise mathematical operations

V = np.array([30, 40, 50, 60])   # speeds (m/s)

# -----------------------------------------
# Lift calculation (vectorized)
# -----------------------------------------
# V**2 is applied to all elements at once
# No for-loop is required

Lift = 0.5 * rho * V**2 * S * CL

# -----------------------------------------
# Output
# -----------------------------------------
print(Lift)
```

Case 7: NumPy Pairwise Calculation (Lift vs Velocity)

```python
# -----------------------------------------
# Import NumPy
# -----------------------------------------
import numpy as np

# -----------------------------------------
# Constants
# -----------------------------------------

S  = 47          # Wing area
CL = 0.5         # Constant lift coefficient

# -----------------------------------------
# NumPy arrays of SAME length
# -----------------------------------------
# Each index represents one flight condition

rho = np.array([1.225, 1.112, 1.007])   # densities
V   = np.array([60, 80, 100])            # speeds (m/s)

# -----------------------------------------
# Pairwise Lift calculation
# -----------------------------------------
# rho[i] is multiplied with V[i] automatically
# No loop is required

Lift = 0.5 * rho * V**2 * S * CL

# -----------------------------------------
# Output
# -----------------------------------------
print(Lift)
```

Case 8: NumPy Broadcasting (All Combinations – Lift Matrix)

```python
# -----------------------------------------
# Import NumPy
# -----------------------------------------
import numpy as np

# -----------------------------------------
# Constants
# -----------------------------------------

S  = 47          # Wing area
CL = 0.5         # Constant lift coefficient

# -----------------------------------------
# NumPy arrays
# -----------------------------------------

rho = np.array([1.225, 1.112, 1.007])   # densities
V   = np.array([60, 80, 100])            # speeds

# -----------------------------------------
# Broadcasting using reshape
# -----------------------------------------
# rho.reshape(3,1) converts:
# [a, b, c] → [[a],
#              [b],
#              [c]]

Lift = 0.5 * rho.reshape(3,1) * V**2 * S * CL

# -----------------------------------------
# Output (matrix)
# -----------------------------------------
print(Lift)
```

Case 9: for loop with NumPy Arrays (Nested Loop, Lift vs Velocity)

```python
# -----------------------------------------
# Import NumPy
# -----------------------------------------
import numpy as np

# -----------------------------------------
# Constants
# -----------------------------------------

S  = 47          # Wing area
CL = 0.5         # Constant lift coefficient

# -----------------------------------------
# NumPy arrays
# -----------------------------------------

rho = np.array([1.225, 1.112, 1.007])   # densities
V   = np.array([60, 80, 100])            # speeds

# -----------------------------------------
# Nested for-loop with NumPy arrays
# -----------------------------------------
# Even though rho and V are NumPy arrays,
# they can still be iterated like Python lists

for r in rho:
    for v in V:

        Lift = 0.5 * r * v**2 * S * CL

        print(f"rho={r} kg/m³, V={v} m/s --> Lift={Lift:.2f} N")
```

```python
# -----------------------------------------
# Import libraries
# -----------------------------------------
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------------
# Constants
# -----------------------------------------

S  = 47
CL = 0.5

# -----------------------------------------
# Arrays
# -----------------------------------------
# Weight array kept for structure consistency
# (not used in Lift equation directly)

Weight = np.array([80000, 103047])     # N
rho    = np.array([1.225, 0.736])      # sea level and high altitude
V      = np.linspace(30, 300, 50)

# -----------------------------------------
# Nested loop for plotting
# -----------------------------------------
# Outer loop → weight cases
# Inner loop → density cases
# Each combination produces one curve

for W in Weight:
    for r in rho:

        Lift = 0.5 * r * V**2 * S * CL

        plt.plot(V, Lift, label=f"rho={r} kg/m³ (W={W} N)")

# -----------------------------------------
# Plot formatting
# -----------------------------------------

plt.xlabel("Speed V (m/s)")
plt.ylabel("Lift (N)")
plt.title("Lift vs Velocity for Multiple Density & Weight Cases")
plt.grid(True)
plt.legend()
plt.show()
```
