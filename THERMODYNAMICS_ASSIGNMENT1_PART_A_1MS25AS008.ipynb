***for loop***

```python
# Case 2: Vary volume, constant mass and temperature (Ideal Gas Law)

# -----------------------------
# Given constants
# -----------------------------

m = 3.0          # mass of air in kg
R = 287          # specific gas constant for air (J/kg·K)
T = 300          # temperature in Kelvin (assumed constant)

# -----------------------------
# Volume range (m^3)
# -----------------------------

V = [0.5, 0.8, 1.0, 1.5, 2.0, 3.0, 4.0, 5.0]

# -----------------------------
# Loop to calculate pressure
# -----------------------------

for v in V:
    # Ideal Gas Law:
    # p = (m * R * T) / V
    p = (m * R * T) / v

    # Display results (similar format to thrust code)
    print(f"Volume = {v:.2f} m^3 --> Pressure = {p:.2f} Pa")
```

```python
# Case 3: Using zip() for element-wise calculation (Ideal Gas Law)
# Pressure calculation for varying volume and mass (aircraft propulsion context)

# -----------------------------
# Given constants
# -----------------------------

R = 287          # Specific gas constant for air (J/kg·K)
T = 300          # Temperature in Kelvin (assumed constant)

# -----------------------------
# Lists of mass and volume
# (Can represent different flow conditions in engine)
# -----------------------------

m = [1.5, 2.0, 2.5, 3.0, 3.5, 4.0]      # Mass of air in kg
V = [0.5, 0.8, 1.0, 1.5, 2.0, 2.5]      # Volume in m^3

# -----------------------------
# Loop using zip()
# -----------------------------

for mass, vol in zip(m, V):
    # Ideal Gas Law:
    # p = (m * R * T) / V
    p = (mass * R * T) / vol

    # Print results (similar format to thrust-required code)
    print(f"Mass = {mass} kg, Volume = {vol} m^3 --> Pressure = {p:.2f} Pa")
```

# Nested ***loop***

```python
# Case 4: Nested loop for full pressure–volume relations
# Ideal Gas Law applied to aircraft propulsion / thermodynamics

# -----------------------------
# Given constants
# -----------------------------

R = 287          # Specific gas constant for air (J/kg·K)
T = 300          # Temperature in Kelvin (assumed constant)

# -----------------------------
# Lists of mass and volume
# (Represent different engine flow conditions)
# -----------------------------

m = [1.0, 1.5, 2.0, 2.5, 3.0]            # Mass of air in kg
V = [0.5, 0.8, 1.0, 1.5, 2.0, 2.5]       # Volume in m^3

# -----------------------------
# Nested loop calculation
# -----------------------------

for mass in m:
    for vol in V:
        # Ideal Gas Law:
        # p = (m * R * T) / V
        p = (mass * R * T) / vol

        # Display results (same style as TR curves)
        print(f"Mass = {mass} kg   Volume = {vol} m^3 --> Pressure = {p:.2f} Pa")
```

***while loop***

```python
# -----------------------------
# Ideal Gas Law using WHILE loop
# Aircraft Propulsion / Thermodynamics
# -----------------------------

# Given constants
m = 2.0          # Mass of air (kg)
R = 287          # Specific gas constant for air (J/kg·K)
T = 300          # Temperature (K)

# Initial volume
V = 10.0         # Volume in m^3

# Pressure limit (threshold)
pressure_limit = 10000  # Pascals (Pa)

# -----------------------------
# WHILE loop
# -----------------------------

while True:
    # Ideal Gas Law
    # p = (m * R * T) / V
    p = (m * R * T) / V

    # Print current volume and pressure
    print(f"Volume = {V:.1f} m^3 --> Pressure = {p:.0f} Pa")

    # Check pressure condition
    if p > pressure_limit:
        print("Pressure limit exceeded!")
        break   # Stop the loop

    # Decrease volume (compression process)
    V -= 1
    # V = V - 1
```

```python
# Case 5: Multiple masses in a loop
# Ideal Gas Law – Pressure vs Volume
# Aircraft Propulsion / Thermodynamics

import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Gas and thermodynamic constants
# -----------------------------

R = 287          # Specific gas constant for air (J/kg·K)
T = 300          # Temperature in Kelvin (assumed constant)

# -----------------------------
# Different mass values (analogous to different densities)
# -----------------------------

m = np.array([1.0, 1.5, 2.0, 2.5, 3.0])   # Mass of air in kg

# -----------------------------
# Volume range
# -----------------------------

V = np.linspace(0.5, 5.0, 50)   # Volume in m^3

# -----------------------------
# Create figure
# -----------------------------

plt.figure(1)

# -----------------------------
# Loop over mass values
# -----------------------------

for mass in m:
    # Ideal Gas Law:
    # p = (m * R * T) / V
    p = (mass * R * T) / V

    # Plot pressure–volume curve
    plt.plot(V, p, linewidth=2, label=f"m = {mass} kg")

# -----------------------------
# Plot formatting
# -----------------------------

plt.xlabel("Volume V (m³)")
plt.ylabel("Pressure p (Pa)")
plt.title("Pressure vs Volume for Different Masses (Ideal Gas Law)")
plt.grid(True)
plt.legend()

plt.show()
```

```python
# Case 6: Nested loop (vary mass and temperature)
# Ideal Gas Law – Pressure vs Volume
# Aircraft Propulsion / Thermodynamics

import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Gas constant
# -----------------------------

R = 287          # Specific gas constant for air (J/kg·K)

# -----------------------------
# Different operating conditions
# (analogous to different weight and density)
# -----------------------------

mass = np.array([1.5, 3.0])        # Mass of air in kg (light & heavy flow)
T = np.array([300, 600])           # Temperature in K (low & high temperature)

# -----------------------------
# Volume range
# -----------------------------

V = np.linspace(0.5, 5.0, 50)       # Volume in m^3

# -----------------------------
# Create figure
# -----------------------------

plt.figure(figsize=(8, 6))

# -----------------------------
# Nested loop calculation
# -----------------------------

for m in mass:
    for temp in T:
        # Ideal Gas Law:
        # p = (m * R * T) / V
        p = (m * R * temp) / V

        # Plot P–V curve
        plt.plot(
            V,
            p,
            linewidth=2,
            label=f"m = {m} kg, T = {temp} K"
        )

# -----------------------------
# Plot formatting
# -----------------------------

plt.xlabel("Volume V (m³)")
plt.ylabel("Pressure p (Pa)")
plt.title("Pressure vs Volume for Multiple Mass and Temperature Conditions")
plt.grid(True)
plt.legend()
plt.show()
```
