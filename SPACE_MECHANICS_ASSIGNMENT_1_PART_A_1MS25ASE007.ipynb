# ***CASE 1 SIMPLE FOR LOOP***

```python
# Case: Constant masses, varying distance

G  = 6.67430e-11          # gravitational constant (N·m^2/kg^2)
m1 = 5.972e24             # mass of Earth (kg)
m2 = 1000                 # mass of satellite (kg)

# Distance from Earth's center (m)
# Includes surface + various orbital altitudes
r = [
    6.371e6,   # Earth surface
    6.771e6,   # ~400 km (LEO - ISS)
    7.371e6,   # ~1000 km
    8.371e6,   # ~2000 km
    1.0371e7,  # ~4000 km
    1.6371e7,  # ~10000 km
    2.6371e7,  # ~20000 km
    4.2164e7   # GEO
]

for dist in r:
    F = G * m1 * m2 / dist**2
    print(f"Distance = {dist:.3e} m --> Gravitational Force = {F:.2f} N")
```

# ***CASE 2 FOR LOOP WITH ZIP***

```python
G  = 6.67430e-11        # gravitational constant (N·m^2/kg^2)
m1 = 5.972e24           # mass of Earth (kg)
m2 = 1000               # mass of satellite (kg)
alt_km = [0, 400, 1000, 2000, 10000, 35786]   # km
r = [
    6.371e6,    # Earth surface
    6.771e6,    # ~400 km (LEO)
    7.371e6,    # ~1000 km
    8.371e6,    # ~2000 km
    1.6371e7,   # ~10000 km
    4.2164e7    # GEO
]

for alt, dist in zip(alt_km, r):
    F = G * m1 * m2 / dist**2
    print(f"Altitude = {alt:6} km, r = {dist:.3e} m --> Force = {F:8.2f} N")
```

# ***CASE 3 LOOPING WITHOUT ZIP***

```python
G  = 6.67430e-11        # gravitational constant (N·m^2/kg^2)
m1 = 5.972e24           # mass of Earth (kg)
m2 = 1000               # mass of satellite (kg)
alt_km = [0, 400, 1000, 2000, 10000, 35786]   # km
r = [
    6.371e6,    # Earth surface
    6.771e6,    # ~400 km (LEO)
    7.371e6,    # ~1000 km
    8.371e6,    # ~2000 km
    1.6371e7,   # ~10000 km
    4.2164e7    # GEO
]
for i in range(len(r)):
    alt  = alt_km[i]
    dist = r[i]

    F = G * m1 * m2 / dist**2

    print(
        f"Altitude = {alt:6} km, "
        f"r = {dist:.3e} m --> "
        f"Gravitational Force = {F:8.2f} N"
    )
```

# ***CASE 4 WITH NESTED LOOP***

```python
G  = 6.67430e-11        # gravitational constant (N·m^2/kg^2)
m1 = 5.972e24           # mass of Earth (kg)
m2 = 1000               # mass of satellite (kg)
# Orbital altitude bands (km)
altitude_bands = {
    "Surface / LEO": [6.371e6, 6.571e6, 6.771e6],
    "MEO":           [1.0371e7, 1.3371e7, 1.6371e7],
    "High / GEO":    [2.6371e7, 3.2164e7, 4.2164e7]
}

for band, r_list in altitude_bands.items():
    print(f"\n--- {band} ---")
    for r in r_list:
        F = G * m1 * m2 / r**2
        print(f"r = {r:.3e} m --> Gravitational Force = {F:8.2f} N")
```

# ***NUMPY***

# ***CASE 1 SINGLE MASSES, MULTIPLE DISTANCES***

```python
# Case 1: Single-mass pair, multiple distances
import numpy as np

# Constants (realistic space mechanics)
G  = 6.67430e-11         # gravitational constant (N·m^2/kg^2)
m1 = 5.972e24            # mass of Earth (kg)
m2 = 1000                # mass of satellite (kg)

# r-values (Earth surface + 3 increasing orbital radii)
r = np.array([
    6.371e6,   # surface
    6.771e6,   # ~400 km
    7.371e6,   # ~1000 km
    8.371e6    # ~2000 km
])

# Newton’s law vectorized
F = G * m1 * m2 / r**2

print(F)
```

# ***CASE 2 PAIRWISE CALCULATION***

```python
# Case 2: Pairwise calculation (r and alt_km same length)
import numpy as np

# Constants (realistic)
G  = 6.67430e-11          # gravitational constant (N·m^2/kg^2)
m1 = 5.972e24             # Earth mass (kg)
m2 = 1000                 # satellite mass (kg)

# paired values – same lengths, just like rho & V
alt_km = np.array([0, 400, 1000])   # altitudes for interpretation (not needed in calc)
r      = np.array([
    6.371e6,   # surface
    6.771e6,   # ~400 km
    7.371e6    # ~1000 km
])

# Newton’s law applied element-wise
F = G * m1 * m2 / r**2

print(F)
```

# ***CASE 3 ALL COMBINATIONS USING BROADCASTING***

```python
# Case 3: All combinations using broadcasting (matrix of gravitational force)
import numpy as np

# constants
G  = 6.67430e-11         # gravitational constant (N·m^2/kg^2)
m1 = 5.972e24            # Earth mass (kg)
m2 = 1000                # satellite mass (kg)

# radii list 1 (rows)
r1 = np.array([
    6.371e6,    # Earth surface
    6.771e6,    # ~400 km (LEO)
    7.371e6     # ~1000 km
])

# radii list 2 (columns)
r2 = np.array([
    8.371e6,     # ~2000 km
    1.6371e7,    # ~10000 km
    4.2164e7     # GEO
])

# broadcasting gravitational formula
F = G * m1 * m2 / (r1.reshape(3,1) * r2.reshape(1,3))

# or equivalently: F = G*m1*m2 / r^2   but storing all pairs
# r^2 computed automatically as r1*r2 in pairwise matrix form

print(F)
```

# ***CASE 4 NESTED LOOP WITH NUMPY ARRAYS***

```python
# Case 4: Nested loop with NumPy arrays (Gravitational Force Table)
import numpy as np

# constants (realistic)
G  = 6.67430e-11         # gravitational constant (N·m^2/kg^2)
m1 = 5.972e24            # Earth mass (kg)
m2 = 1000                # satellite mass (kg)

# radii arrays (distance from center of Earth)
# outer + inner loop patterns like rho and V
r_outer = np.array([6.371e6, 6.771e6, 7.371e6])   # surface + altitudes
r_inner = np.array([8.371e6, 1.6371e7, 4.2164e7]) # higher altitudes

for r1 in r_outer:
    for r2 in r_inner:
        # Using pairwise equivalent radius
        # Such as computing forces for combinations of radii
        r = r2     # could choose r1 or r2 depending on modeling need

        F = G * m1 * m2 / r**2

        print(f"r_outer={r1:.2e} m, r_inner={r2:.2e} m --> F={F:.2f} N")
```

# ***MATPLOTLIB***

# ***CASE 1 SIMPLE LINE PLOT (FORCE VS DISTANCE)***

```python
import numpy as np
import matplotlib.pyplot as plt

# Physical constants
G  = 6.67430e-11         # gravitational constant (N·m^2/kg^2)
m1 = 5.972e24            # Earth mass (kg)
m2 = 1000                # satellite mass (kg)

# radius range (meters) – center of Earth to distances in space
r = np.linspace(6.371e6, 4.2164e7, 200)   # from surface to GEO

# gravitational force
F = G * m1 * m2 / r**2

# figure 1: simple gravitational force vs radius
plt.figure(1)
plt.plot(r, F)  # no explicit color (per instructions)
plt.xlabel("Distance r (m)")
plt.ylabel("Gravitational Force F (N)")
plt.title("Gravitational Force vs Distance (Newton’s Law)")
plt.grid(True)
plt.show()
```

# ***CASE 2 TWO SEPARATE FIGURES (LINE + SCATTER)***

```python
# Case 2: Two separate figures (line + scatter) – Newton's gravitation

import numpy as np
import matplotlib.pyplot as plt

# Physical constants
G  = 6.67430e-11        # gravitational constant (N·m^2/kg^2)
m1 = 5.972e24           # Earth mass (kg)
m2 = 1000               # satellite mass (kg)

# Distance vector (from Earth's center to GEO)
r = np.linspace(6.371e6, 4.2164e7, 50)

# Gravitational force
F = G * m1 * m2 / r**2

# Figure 1: Line plot
plt.figure(1)
plt.plot(r, F, linewidth=2)
plt.xlabel("Distance r (m)")
plt.ylabel("Gravitational Force F (N)")
plt.title("Gravitational Force vs Distance (Line)")
plt.grid(True)

# Figure 2: Scatter plot
plt.figure(2)
plt.scatter(r, F, s=50)
plt.xlabel("Distance r (m)")
plt.ylabel("Gravitational Force F (N)")
plt.title("Gravitational Force vs Distance (Scatter)")
plt.grid(True)

plt.show()
```

# ***CASE 3 SUBPLOTS (SIDE BY SIDE)***

```python
# Case 3: Subplots (side by side) – Newton's Gravitational Force
import numpy as np
import matplotlib.pyplot as plt

# Physical constants
G  = 6.67430e-11         # gravitational constant (N·m^2/kg^2)
m1 = 5.972e24            # Earth mass (kg)
m2 = 1000                # satellite mass (kg)

# Distance range from Earth's center
r = np.linspace(6.371e6, 4.2164e7, 50)   # from surface to GEO

# Gravitational force from Newton's law
F = G * m1 * m2 / r**2

plt.figure(figsize=(12, 5))

# Left subplot – line
plt.subplot(1, 2, 1)
plt.plot(r, F, linewidth=2)
plt.xlabel("Distance r (m)")
plt.ylabel("Force F (N)")
plt.title("Gravitational Force vs Distance (Line)")
plt.grid(True)

# Right subplot – scatter
plt.subplot(1, 2, 2)
plt.scatter(r, F, s=50)
plt.xlabel("Distance r (m)")
plt.ylabel("Force F (N)")
plt.title("Gravitational Force vs Distance (Scatter)")
plt.grid(True)

plt.show()
```

# ***CASE 4 OVERLAY PLOT***

```python
# Case 4: Overlay plot – Newton's Law of Gravitation

import numpy as np
import matplotlib.pyplot as plt

# gravitational constants
G  = 6.67430e-11         # gravitational constant (N·m^2/kg^2)
m1 = 5.972e24            # Earth mass (kg)

# varying distance (analogous to V array)
r = np.linspace(6.371e6, 4.2164e7, 50)   # Earth radius to GEO

# First condition: small satellite
m2_1 = 500      # kg
F1 = G * m1 * m2_1 / r**2

plt.plot(r, F1, linewidth=2, label="Satellite mass = 500 kg")
plt.scatter(r, F1, s=30)

# Second condition: heavier satellite
m2_2 = 1500     # kg
F2 = G * m1 * m2_2 / r**2

plt.plot(r, F2, linewidth=2, label="Satellite mass = 1500 kg")

# formatting
plt.xlabel("Distance r (m)")
plt.ylabel("Gravitational Force F (N)")
plt.title("Gravitational Force vs Distance (Overlay for two satellite masses)")
plt.grid(True)
plt.legend()
plt.show()
```
