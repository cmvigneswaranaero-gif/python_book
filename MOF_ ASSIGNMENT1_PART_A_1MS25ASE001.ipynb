# ***FOR LOOP***

case 1: basic for loop

```python
# Case 1: Pressure variation with height (Hydrostatic equation)

# Import math module (not strictly needed here, but good practice)
import math

# Atmospheric pressure at reference level (Pa)
Pa = 101325       # Standard atmospheric pressure at sea level

# Fluid density (kg/m^3)
rho = 1000        # Density of water (use 1.225 for air)

# Acceleration due to gravity (m/s^2)
g = 9.81

# List of heights at which pressure is to be calculated (m)
h_list = [0, 5, 10, 20, 30, 40, 50, 75, 100]

# Print table heading
print("Height (m) | Pressure (Pa)")
print("----------------------------")

# Loop through each height in the list
for h in h_list:

    # Calculate pressure at height h using hydrostatic equation
    P = Pa + rho * g * h

    # Print height and corresponding pressure in formatted form
    print(f"{h:10.1f} | {P:12.2f}")
```

case 2: Looping over two list with zip()

```python
# Case 2: Using zip() for element-wise pressure calculation
# Equation: P = Pa + rho * g * h

# Atmospheric pressure at reference level (Pa)
Pa = 101325    # Standard atmospheric pressure at sea level

# Acceleration due to gravity (m/s^2)
g = 9.81

# List of heights (m) at different locations
h = [0, 500, 1000, 1500, 2000, 2500]

# Corresponding list of air densities (kg/m^3)
# (Density decreases with altitude)
rho = [1.225, 1.112, 1.007, 0.909, 0.819, 0.736]

# Print heading for formatted output
print("Height (m) | Density (kg/m^3) | Pressure (Pa)")
print("------------------------------------------------")

# Loop through height and density simultaneously using zip()
for height, r in zip(h, rho):

    # Calculate pressure using hydrostatic equation
    P = Pa + r * g * height

    # Print height, density, and pressure in formatted form
    print(f"{height:10.0f} | {r:17.3f} | {P:13.2f}")
```

case 3: Looping without zip()

```python
# Case 3: Index-based looping for pressure calculation
# Equation: P = Pa + rho * g * h

# Atmospheric pressure at reference level (Pa)
Pa = 101325     # Standard atmospheric pressure at sea level

# Acceleration due to gravity (m/s^2)
g = 9.81

# List of heights at different locations (m)
h = [0, 500, 1000, 1500, 2000, 2500]

# Corresponding list of fluid densities (kg/m^3)
rho = [1.225, 1.112, 1.007, 0.909, 0.819, 0.736]

# Print heading for output table
print("Height (m) | Density (kg/m^3) | Pressure (Pa)")
print("------------------------------------------------")

# Loop over the index of the list
for i in range(len(h)):

    # Extract height using index
    height = h[i]

    # Extract corresponding density using index
    r = rho[i]

    # Calculate pressure using hydrostatic equation
    P = Pa + r * g * height

    # Print height, density, and pressure in formatted form
    print(f"{height:10.0f} | {r:17.3f} | {P:13.2f}")
```

case 4: Nested loop

```python
# Case 4: Nested loop for pressure calculations
# Equation: P = Pa + rho * g * h

# Atmospheric pressure at reference level (Pa)
Pa = 101325

# Acceleration due to gravity (m/s^2)
g = 9.81

# List of densities (kg/m^3)
rho_list = [1.225, 1.112, 1.007, 0.909, 0.819, 0.736]

# List of heights (m)
h_list = [0, 500, 1000, 1500, 2000, 2500]

# Print table header
print("Density (kg/m^3) | Height (m) | Pressure (Pa)")
print("------------------------------------------------")

# Outer loop: iterate over densities
for r in rho_list:

    # Inner loop: iterate over heights
    for h in h_list:

        # Calculate pressure using hydrostatic equation
        P = Pa + r * g * h

        # Print density, height, and corresponding pressure
        print(f"{r:17.3f} | {h:10.0f} | {P:12.2f}")
```

# ***NUMPY***

case 1: Single density, multiple heights using NumPy

```python
# Case 1: Single density, multiple heights using NumPy
import numpy as np  # Import NumPy for array operations

# ---------------- Constants ----------------
Pa = 101325       # Reference atmospheric pressure at sea level (Pa)
rho = 1.225       # Air density at sea level (kg/m^3)
g = 9.81          # Acceleration due to gravity (m/s^2)

# Heights at which pressure is calculated (m)
h = np.array([0, 100, 200, 500, 1000, 1500, 2000])

# ---------------- Calculations ----------------
# Hydrostatic pressure at each height
P = Pa + rho * g * h

# ---------------- Output ----------------
print("Heights (m):", h)
print("Density (kg/m^3):", rho)
print("Pressure (Pa):", P)
```

case 2: Pairwise calculation using NumPy

```python
# Case 2: Pairwise calculation using NumPy
# Equation: P = Pa + rho * g * h

import numpy as np  # Import NumPy for numerical array operations

# ---------------- Constants ----------------
Pa = 101325        # Atmospheric pressure at sea level (Pa)
g = 9.81           # Acceleration due to gravity (m/s^2)

# ---------------- Input Arrays ----------------
# Density values (kg/m^3)
rho = np.array([1.225, 1.112, 1.007])

# Corresponding heights (m)
# Each height matches one density (pairwise calculation)
h = np.array([0, 500, 1000])

# ---------------- Calculations ----------------
# Pressure calculation using hydrostatic equation
P = Pa + rho * g * h

# ---------------- Output ----------------
print("Density (kg/m^3):", rho)
print("Height (m):", h)
print("Pressure (Pa):", P)
```

case 3: All combinations using broadcasting (matrix pressure table)

```python
# Case 3: All combinations using broadcasting (matrix pressure table)
# Equation: P = Pa + rho * g * h

import numpy as np  # Import NumPy for vectorized numerical operations

# ---------------- Constants ----------------
Pa = 101325        # Atmospheric pressure at sea level (Pa)
g = 9.81           # Acceleration due to gravity (m/s^2)

# ---------------- Input Arrays ----------------
# Densities of fluid at different conditions (kg/m^3)
rho = np.array([1.225, 1.112, 1.007])

# Heights at which pressure is calculated (m)
h = np.array([0, 500, 1000, 1500])

# ---------------- Broadcasting Preparation ----------------
# Reshape density array into a column vector (3x1)
rho_col = rho.reshape(len(rho), 1)

# ---------------- Calculations ----------------
# Pressure matrix using NumPy broadcasting
# Each row corresponds to one density
# Each column corresponds to one height
P = Pa + rho_col * g * h

# ---------------- Output ----------------
print("Pressure Matrix (Pa):")
print(P)
```

Case 4: Nested loop with NumPy arrays for pressure calculation

```python
# Case 4: Nested loop with NumPy arrays for pressure calculation
# Equation: P = Pa + rho * g * h

import numpy as np  # Import NumPy for numerical operations

# ---------------- Constants ----------------
Pa = 101325        # Atmospheric pressure at sea level (Pa)
g = 9.81           # Acceleration due to gravity (m/s^2)

# ---------------- Input Arrays ----------------
# Densities of air at different altitudes (kg/m^3)
rho = np.array([1.225, 1.112, 1.007])

# Heights at which pressure is calculated (m)
h = np.array([0, 500, 1000])

# ---------------- Nested Loop Calculation ----------------
# Outer loop: iterate over density values
for r in rho:

    # Inner loop: iterate over height values
    for height in h:

        # Calculate pressure using hydrostatic equation
        P = Pa + r * g * height

        # Print results in a clear, readable format
        print(f"rho = {r} kg/m³, h = {height} m --> Pressure = {P:.2f} Pa")
```

# *MATPLOTLIB*

case 1: Simple line plot (Pressure vs Height)

```python
# Case 1: Simple line plot (Pressure vs Height)
# Equation used: P = Pa + rho * g * h

import numpy as np
import matplotlib.pyplot as plt

# ---------------- Physical Parameters ----------------
Pa = 101325        # Atmospheric pressure at sea level (Pa)
rho = 1.225        # Air density at sea level (kg/m^3)
g = 9.81           # Acceleration due to gravity (m/s^2)

# ---------------- Height Range ----------------
# Heights from 0 to 5000 m (50 points)
h = np.linspace(0, 5000, 50)

# ---------------- Pressure Calculation ----------------
# Hydrostatic pressure at each height
P = Pa + rho * g * h

# ---------------- Figure 1: Line Plot ----------------
plt.figure(1)                          # Create a new figure
plt.plot(h, P, linewidth=2)            # Plot pressure vs height
plt.xlabel("Height h (m)")             # X-axis label
plt.ylabel("Pressure P (Pa)")          # Y-axis label
plt.title("Pressure vs Height (Line)") # Plot title
plt.grid(True)                         # Enable grid
plt.show()                             # Display the plot
```

case 4: Overlay plot- Pressure vs Height for two different densities

```python
# Overlay plot: Pressure vs Height for two different densities
# Equation used: P = Pa + rho * g * h

import numpy as np
import matplotlib.pyplot as plt

# ---------------- Physical Constants ----------------
Pa = 101325        # Atmospheric pressure at sea level (Pa)
g = 9.81           # Acceleration due to gravity (m/s^2)

# ---------------- Height Range ----------------
# Heights from 0 to 5000 m
h = np.linspace(0, 5000, 50)

# ---------------- First Condition: Sea Level ----------------
rho1 = 1.225       # Density at sea level (kg/m^3)

# Pressure calculation for first density
P1 = Pa + rho1 * g * h

# Plot line and scatter for first density
plt.plot(h, P1, linewidth=2, label="rho = 1.225 kg/m³")
plt.scatter(h, P1, s=30)

# ---------------- Second Condition: Higher Altitude ----------------
rho2 = 0.909       # Density at higher altitude (kg/m^3)

# Pressure calculation for second density
P2 = Pa + rho2 * g * h

# Plot line for second density
plt.plot(h, P2, linewidth=2, label="rho = 0.909 kg/m³")

# ---------------- Plot Formatting ----------------
plt.xlabel("Height h (m)")                          # X-axis label
plt.ylabel("Pressure P (Pa)")                       # Y-axis label
plt.title("Pressure vs Height (Overlay for Two Densities)")
plt.grid(True)                                      # Enable grid
plt.legend()                                        # Show legend
plt.show()                                          # Display plot
```
