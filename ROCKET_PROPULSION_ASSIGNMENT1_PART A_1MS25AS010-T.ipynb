```python
#CASE-1
import math

# Constants
g0 = 9.80665   # m/s^2, standard gravity

# Inputs
isp = 300      # specific impulse in seconds
m0 = 50000     # initial mass (kg)
mf_values = [40000, 35000, 30000, 25000, 20000, 15000, 10000]  # final masses to vary

# Loop over different final masses
for mf in mf_values:
    dv = isp * g0 * math.log(m0 / mf)
    print(f"Initial Mass = {m0} kg, Final Mass = {mf} kg --> Δv = {dv:.2f} m/s")
```

Looping over two lists with zip()

```python
#CASE-2 zip() for element-wise calculation
import math

# Constants
g0 = 9.80665   # m/s^2, standard gravity
isp = 300      # specific impulse in seconds

# Lists of initial and final masses
m0_list = [50000, 60000, 70000, 80000, 90000, 100000]   # initial masses (kg)
mf_list = [20000, 25000, 30000, 35000, 40000, 45000]    # final masses (kg)

# Loop through both lists simultaneously using zip()
for m0, mf in zip(m0_list, mf_list):
    dv = isp * g0 * math.log(m0 / mf)
    print(f"Initial Mass = {m0} kg, Final Mass = {mf} kg --> Δv = {dv:.2f} m/s")

# Why use zip():
# zip() lets us iterate over two lists element-wise in parallel.
# Each initial mass is paired with its corresponding final mass,
# ensuring calculations are matched correctly without manual indexing.
```

CASE-3 Index-based looping

```python
#CASE-3 Index-based looping
import math

# Constants
g0 = 9.80665   # m/s^2, standard gravity
isp = 300      # specific impulse in seconds

# Lists of initial and final masses
m0_list = [50000, 60000, 70000, 80000, 90000, 100000]   # initial masses (kg)
mf_list = [20000, 25000, 30000, 35000, 40000, 45000]    # final masses (kg)

# Index-based looping
for i in range(len(m0_list)):
    m0 = m0_list[i]
    mf = mf_list[i]
    dv = isp * g0 * math.log(m0 / mf)
    print(f"Initial Mass = {m0} kg, Final Mass = {mf} kg --> Δv = {dv:.2f} m/s")

# Why use index-based looping:
# This method gives direct control over list positions.
# It is useful when you need to access multiple lists by index,
# or when you plan to manipulate indices (e.g., skipping, slicing).
```

 Case 4: Nested loop

```python
# Case 4: Nested loop
import math

# Constants
g0 = 9.80665   # m/s^2, standard gravity

# Lists of specific impulse values and mass ratios
isp_list = [250, 300, 350, 400]                 # Isp values in seconds
mass_pairs = [(50000, 20000), (60000, 25000),   # (m0, mf) pairs
              (70000, 30000), (80000, 35000),
              (90000, 40000)]

# Nested loop: iterate over each Isp and each mass pair
for isp in isp_list:
    for m0, mf in mass_pairs:
        dv = isp * g0 * math.log(m0 / mf)
        print(f"Isp = {isp} s, Initial Mass = {m0} kg, Final Mass = {mf} kg --> Δv = {dv:.2f} m/s")

# Why use nested loops:
# Nested loops let us explore combinations of two variables (here Isp and mass ratio).
# This is useful when building full Δv curves across multiple propellant types and mass scenarios.
```

 Case 1: Rocket equation with multiple mass pairs

```python
# Case 1: Rocket equation with multiple mass pairs
import numpy as np

# Constants
g0 = 9.80665   # m/s^2, standard gravity
isp = 300      # specific impulse in seconds

# Mass arrays (initial and final masses for rockets)
m0 = np.array([50000, 60000, 70000, 80000])   # initial masses (kg)
mf = np.array([20000, 25000, 30000, 35000])   # final masses (kg)

# Rocket equation using NumPy (vectorized Δv calculation)
dv = isp * g0 * np.log(m0 / mf)

print(dv)
```

# Case 2: Pairwise calculation

```python
# Case 2: Pairwise calculation (ρ and V same length)
import numpy as np

# Constants
g0 = 9.80665   # m/s^2, standard gravity
isp = 300      # specific impulse in seconds

# Pairwise arrays (same length)
m0 = np.array([50000, 60000, 70000])   # initial masses (kg)
mf = np.array([20000, 25000, 30000])   # final masses (kg)

# Rocket equation (vectorized pairwise calculation)
dv = isp * g0 * np.log(m0 / mf)

print(dv)

# Why use this method:
# 1. Pairwise calculation ensures each initial mass (m0[i]) is matched with its corresponding final mass (mf[i]).
# 2. Using NumPy arrays allows vectorized operations, so the formula is applied to all pairs at once.
# 3. This avoids writing explicit loops, making the code shorter, cleaner, and easier to read.
# 4. NumPy is optimized for performance, so calculations run faster, especially with large datasets.
# 5. It reduces the chance of indexing errors compared to manual looping.
```

 Case 3: All combinations using broadcasting

```python
 # Case 3: All combinations using broadcasting
import numpy as np

# Constants
g0 = 9.80665   # m/s^2, standard gravity

# Arrays of specific impulse values and mass ratios
isp = np.array([250, 300, 350])                 # Isp values (s)
m0  = np.array([50000, 60000, 70000])           # initial masses (kg)
mf  = np.array([20000, 25000, 30000])           # final masses (kg)

# Reshape for broadcasting
# isp as column vector (3x1), mass ratio as row vector (1x3)
dv = isp.reshape(3,1) * g0 * np.log(m0 / mf)

print(dv)

# Why use broadcasting:
# 1. Broadcasting lets us compute all combinations of Isp and mass ratios in one step.
# 2. No explicit loops are needed — NumPy automatically expands dimensions to align arrays.
# 3. Produces a clean matrix output where rows correspond to Isp values and columns to mass ratios.
# 4. This method is concise, efficient, and ideal for generating performance tables or Δv curves.
```

 Case 4: for loop with NumPy arrays

```python
# Case 4: for loop with NumPy arrays
import numpy as np
import math

# Constants
g0 = 9.80665   # m/s^2, standard gravity

# Arrays of Isp values and mass pairs
isp = np.array([250, 300, 350])                 # specific impulse (s)
m0  = np.array([50000, 60000, 70000])           # initial masses (kg)
mf  = np.array([20000, 25000, 30000])           # final masses (kg)

# Nested loop over both arrays
for i in isp:
    for j, k in zip(m0, mf):
        dv = i * g0 * math.log(j / k)
        print(f"Isp = {i} s, Initial Mass = {j} kg, Final Mass = {k} kg --> Δv = {dv:.2f} m/s")

# Why use nested loops with NumPy arrays:
# 1. Nested loops let us explore all combinations of two parameters (here Isp and mass ratio).
# 2. Using NumPy arrays keeps the data organized and easy to extend.
# 3. f-strings provide human-readable, formatted output for each combination.
# 4. This method is clear and explicit, making it easy to understand the calculation flow.
```

# Case 1: Simple line plot (Delta-v vs Mass Ratio )

```python

# Case 1: Simple line plot (Delta-v vs Mass Ratio )
import numpy as np
import matplotlib.pyplot as plt

# Constants
g0 = 9.80665   # m/s^2, standard gravity
isp = 300      # specific impulse (s)

# Mass ratio range (m0/mf)
mass_ratio = np.linspace(1.1, 5.0, 50)   # from 1.1 to 5.0

# Rocket equation calculation
dv = isp * g0 * np.log(mass_ratio)

# Figure 1: Line plot
plt.figure(1)
plt.plot(mass_ratio, dv, color='red', linewidth=2)
plt.xlabel("Mass Ratio (m0/mf)")
plt.ylabel("Delta-v (m/s)")
plt.title("Delta-v vs Mass Ratio (Rocket Equation)")
plt.grid(True)
plt.show()

# Why use matplotlib here:
# 1. Matplotlib provides clear visualization of how Δv changes with mass ratio.
# 2. A line plot makes trends easy to interpret (e.g., exponential growth of Δv).
# 3. Visuals are more intuitive than raw numbers, especially for performance analysis.
# 4. Helps compare rocket designs and propellant efficiency in a straightforward way.
```

# Case 6: Nested loop

```python
# Case 6: Nested loop
import numpy as np
import matplotlib.pyplot as plt

# Constants
g0 = 9.80665   # m/s^2, standard gravity

# Two initial masses (light and heavy configurations)
m0_values = np.array([50000, 100000])   # kg
mf = 20000                              # final mass (kg)

# Two Isp values (sea-level vs vacuum engines)
isp_values = np.array([300, 450])       # seconds

# Mass ratio range
mass_ratio = np.linspace(1.1, 5.0, 50)

# Nested loop: vary initial mass and Isp
for m0 in m0_values:
    for isp in isp_values:
        dv = isp * g0 * np.log(m0 / mf)
        plt.plot(mass_ratio, isp * g0 * np.log(mass_ratio),
                 label=f"m0={m0} kg, Isp={isp} s")

# Labels and formatting
plt.xlabel("Mass Ratio (m0/mf)")
plt.ylabel("Delta-v (m/s)")
plt.title("Delta-v vs Mass Ratio for multiple m0 and Isp")
plt.grid(True)
plt.legend()
plt.show()

# Why use nested loops:
# 1. Nested loops allow us to explore combinations of two parameters (here initial mass and Isp).
# 2. Each curve shows how Δv changes with mass ratio for a specific rocket configuration.
# 3. Overlaying them highlights trade-offs between heavier rockets and higher efficiency engines.
# 4. This method is flexible: just add more values to extend comparisons.
```
