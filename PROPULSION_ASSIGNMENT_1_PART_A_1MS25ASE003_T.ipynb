# ***FOR LOOP***

**BASIC FOR LOOP**

```python
# Case 1: Jet propulsion thrust calculation (constant density)
# Thrust = rho * A_inlet * Vi * (Vj - Vi)

# Given parameters
rho = 1.225                  # Air density at sea level (kg/m^3)
A_inlet = 1.2                # Engine inlet area (m^2)
Vi = 60                      # Flight / inlet velocity (m/s)

# Jet exhaust velocities (Vj) in m/s
Vj_list = [80, 100, 120, 140, 160, 180, 200, 220, 240]

# Loop over jet velocities
for Vj in Vj_list:

    # Momentum thrust equation for a simple jet engine
    # Assumes:
    # - Constant density
    # - Fully expanded exhaust (no pressure thrust)
    # - Uniform inlet and exit velocities
    Thrust = rho * A_inlet * Vi * (Vj - Vi)

    # Output results
    print(f"Jet Velocity = {Vj} m/s --> Thrust = {Thrust:.2f} N")
```

# **NESTED LOOP**

```python
# Case 1: Nested loop for full thrust curves (jet momentum theory)

# Air densities at different altitudes (kg/m^3)
rho = [1.225, 1.112, 1.007, 0.909, 0.819, 0.736]

# Jet exhaust velocities (m/s)
Vj = [100, 120, 140, 160, 180, 200, 220, 240,
      260, 280, 300]

# Constants
A_inlet = 1.2     # Engine inlet area (m^2)
Vi = 60           # Inlet / flight velocity (m/s)

# Nested loop (same structure as your TR code)
for r in rho:
    for vj in Vj:

        # Jet momentum thrust equation
        Thrust = r * A_inlet * Vi * (vj - Vi)

        print(f"Density = {r} kg/m^3   Jet Speed = {vj} m/s --> Thrust = {Thrust:.2f} N")
```

**WHILE LOOP — Incrementing jet velocity**

```python
# Case 3: WHILE loop using jet thrust equation

# Constants
rho = 1.225
A_inlet = 1.2
Vi = 60              # Constant inlet velocity

# Initial jet velocity
Vj = 80

# Jet velocity limit
Vj_max = 240

# While loop
while Vj <= Vj_max:

    # Jet momentum thrust
    Thrust = rho * A_inlet * Vi * (Vj - Vi)

    print(f"Vj = {Vj} m/s --> Thrust = {Thrust:.2f} N")

    Vj += 20   # Increment jet velocity
```

# ***NUMPY***

```python
# Case 1: Single density, multiple jet velocities
import numpy as np

# Constants
rho = 1.225                 # Air density (kg/m^3)
A_inlet = 1.2               # Engine inlet area (m^2)
Vi = 60                     # Flight / inlet velocity (m/s)

# Jet velocities (m/s)
Vj = np.array([100, 120, 140, 160])

# Thrust calculation (vectorized)
Thrust = rho * A_inlet * Vi * (Vj - Vi)

print(Thrust)
```

```python
# Case 2: Pairwise thrust calculation (rho and Vj same length)
import numpy as np

# Constants
A_inlet = 1.2
Vi = 60

# Density and jet velocity arrays
rho = np.array([1.225, 1.112, 1.007])   # kg/m^3
Vj  = np.array([140, 180, 220])         # m/s

# Pairwise thrust calculation
Thrust = rho * A_inlet * Vi * (Vj - Vi)

print(Thrust)
```

```python
# Case 3: All combinations using NumPy broadcasting
import numpy as np

# Constants
A_inlet = 1.2
Vi = 60

# Density and jet velocity arrays
rho = np.array([1.225, 1.112, 1.007])   # kg/m^3
Vj  = np.array([100, 140, 180, 220])    # m/s

# Reshape density for broadcasting
Thrust = rho.reshape(3,1) * A_inlet * Vi * (Vj - Vi)

print(Thrust)
```

```python
# Case 4: Nested FOR loop for thrust calculation
import numpy as np

# Constants
A_inlet = 1.2
Vi = 60

# Arrays
rho = np.array([1.225, 1.112, 1.007])
Vj  = np.array([100, 140, 180, 220])

# Nested loop (same style as TR code)
for r in rho:
    for vj in Vj:

        # Jet momentum thrust
        Thrust = r * A_inlet * Vi * (vj - Vi)

        print(f"rho = {r} kg/m³, Vj = {vj} m/s --> Thrust = {Thrust:.2f} N")
```

# ***MATPLOT LIB***

```python
# Case 1: Simple line plot of thrust vs jet velocity (Vj)
# Physics used: Jet momentum theory

import numpy as np
import matplotlib.pyplot as plt

# ------------------ Constants ------------------
rho = 1.225          # Air density at sea level (kg/m^3)
A_inlet = 1.2        # Engine inlet area (m^2)
Vi = 60              # Aircraft / inlet velocity (m/s)

# ------------------ Jet velocity range ------------------
Vj = np.linspace(80, 300, 50)   # Jet exhaust velocity (m/s)

# ------------------ Thrust calculation ------------------
# T = rho * A_inlet * Vi * (Vj - Vi)
T = rho * A_inlet * Vi * (Vj - Vi)

# ------------------ Plot ------------------
plt.figure(1)
plt.plot(Vj, T, linewidth=2)
plt.xlabel("Jet Velocity Vj (m/s)")
plt.ylabel("Thrust T (N)")
plt.title("Thrust vs Jet Velocity (Sea Level)")
plt.grid(True)
plt.show()
```

```python
# Case 4: Overlay plot showing effect of air density on thrust

import numpy as np
import matplotlib.pyplot as plt

# Constants
A_inlet = 1.2
Vi = 60
Vj = np.linspace(80, 300, 50)

# -------- Sea level condition --------
rho1 = 1.225
T1 = rho1 * A_inlet * Vi * (Vj - Vi)
plt.plot(Vj, T1, linewidth=2, label="rho = 1.225 kg/m³")
plt.scatter(Vj, T1, s=30)

# -------- Higher altitude condition --------
rho2 = 0.909
T2 = rho2 * A_inlet * Vi * (Vj - Vi)
plt.plot(Vj, T2, linewidth=2, label="rho = 0.909 kg/m³")

# Plot formatting
plt.xlabel("Jet Velocity Vj (m/s)")
plt.ylabel("Thrust T (N)")
plt.title("Thrust vs Jet Velocity for Two Densities")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Case 5: Thrust vs Vj for multiple densities using FOR loop

import numpy as np
import matplotlib.pyplot as plt

# Constants
A_inlet = 1.2
Vi = 60
Vj = np.linspace(80, 300, 50)

# Different air densities (altitude effect)
rho_list = np.array([1.225, 1.112, 1.007, 0.909, 0.819, 0.736])

plt.figure(1)

# Loop over densities
for rho in rho_list:
    T = rho * A_inlet * Vi * (Vj - Vi)
    plt.plot(Vj, T, linewidth=2, label=f"rho = {rho} kg/m³")

# Plot formatting
plt.xlabel("Jet Velocity Vj (m/s)")
plt.ylabel("Thrust T (N)")
plt.title("Thrust vs Jet Velocity for Different Densities")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Case 6: Nested loop – effect of inlet velocity and density on thrust

import numpy as np
import matplotlib.pyplot as plt

# Constants
A_inlet = 1.2
Vj = np.linspace(80, 300, 50)

# Different flight speeds and densities
Vi_list = np.array([40, 60])          # Inlet velocities (m/s)
rho_list = np.array([1.225, 0.736])   # Sea level and high altitude

# Nested loop
for Vi in Vi_list:
    for rho in rho_list:
        T = rho * A_inlet * Vi * (Vj - Vi)
        plt.plot(Vj, T, label=f"Vi={Vi} m/s, rho={rho} kg/m³")

# Plot formatting
plt.xlabel("Jet Velocity Vj (m/s)")
plt.ylabel("Thrust T (N)")
plt.title("Thrust vs Jet Velocity (Multiple Vi & rho)")
plt.grid(True)
plt.legend()
plt.show()
```
