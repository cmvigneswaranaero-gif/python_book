***for*** ***loop***

```python
# Case1: Cantilever beam with point load at free end
# Vary load W, constant E, I, and L

# Given beam properties
L = 2.5          # Length of beam (m)
E = 200e9        # Young's modulus (Pa) – Steel
I = 8e-6         # Moment of inertia (m^4)

# Load values (N)
W = [500, 1000, 1500, 2000, 2500, 3000]

# Loop through loads
for w in W:
    delta_max = (w * L**3) / (3 * E * I)
    print(f"Load = {w} N --> Maximum Deflection = {delta_max:.6e} m")
```

```python
# Case 2: zip() for element-wise calculation
# Cantilever beam – point load at free end

# Constant beam properties
L = 2.5          # Length of beam (m)
I = 7e-6         # Moment of inertia (m^4)

# Varying load (N) and material property (Pa)
W = [500, 1000, 1500, 2000, 2500, 3000]
E = [200e9, 190e9, 180e9, 170e9, 160e9, 150e9]

# zip() for element-wise pairing
for w, e in zip(W, E):
    delta_max = (w * L**3) / (3 * e * I)
    print(f"Load = {w} N, E = {e:.2e} Pa --> Max Deflection = {delta_max:.6e} m")
```

***Looping Without zip()***

```python
# Case 3: Index-based looping
# Cantilever beam with point load at free end

# Constant beam properties
L = 2.5          # Length of beam (m)
I = 6e-6         # Moment of inertia (m^4)

# Load (N) and Young's modulus (Pa)
W = [500, 1000, 1500, 2000, 2500, 3000]
E = [200e9, 190e9, 180e9, 170e9, 160e9, 150e9]

# Index-based loop
for i in range(len(W)):
    w = W[i]
    e = E[i]

    delta_max = (w * L**3) / (3 * e * I)

    print(f"Load = {w} N, E = {e:.2e} Pa --> Max Deflection = {delta_max:.6e} m")
```

***NESTED LOOP***

```python
# Case 4: Nested loop for full deflection curves
# Cantilever beam – point load at free end

# Constant beam properties
I = 6e-6         # Moment of inertia (m^4)

# Varying beam length (m) and Young's modulus (Pa)
L = [1.5, 2.0, 2.5, 3.0]
E = [200e9, 180e9, 160e9]

# Load values (N)
W = [500, 1000, 1500, 2000]

# Nested loop (similar to rho–V sweep)
for e in E:
    for l in L:
        for w in W:
            delta_max = (w * l**3) / (3 * e * I)
            print(
                f"E = {e:.2e} Pa | Length = {l} m | Load = {w} N "
                f"--> Max Deflection = {delta_max:.6e} m"
            )
```

***WHILE LOOP***

```python
# Constants (beam properties)
L = 2.5          # Beam length (m)
E = 200e9        # Young's modulus (Pa)
I = 6e-6         # Moment of inertia (m^4)

# Initial load
W = 500          # N

# Deflection threshold (m)
threshold = 0.005   # 5 mm

# While loop
while True:
    delta = (W * L**3) / (3 * E * I)
    print(f"Load = {W} N --> Deflection = {delta:.6f} m")

    if delta > threshold:
        print("Deflection exceeded allowable limit!")
        break  # exit the loop

    W += 250   # increase load each iteration
    # W = W + 250
```

```python
# Constants (beam properties)
L = 2.0          # Beam length (m)
E = 200e9        # Young's modulus (Pa)
I = 5e-6         # Moment of inertia (m^4)

# Initial load
W = 100          # N

# Allowable deflection limit (m)
limit = 0.002    # 2 mm

# While loop
while True:
    delta = (W * L**3) / (3 * E * I)
    print(f"Load = {W} N --> Deflection = {delta:.6f} m")

    if delta > limit:
        print("Beam deflection has exceeded allowable limit!")
        break  # stop the loop

    W += 100   # increment load
```

```python
# Case 1: Cantilever beam - Single beam, multiple loads
import numpy as np

# Beam constants
L = 2           # Length of the beam in meters (m)
E = 200e9       # Young's Modulus in Pascals (Pa)
I = 0.0001      # Moment of inertia in m^4

# Loads (N) - multiple cases like multiple speeds in aircraft example
W = np.array([500, 1000, 1500, 2000])

# Maximum deflection formula: delta_max = (W * L^3) / (3 * E * I)
delta_max = (W * L**3) / (3 * E * I)

# Display the results
print("Maximum deflection for different loads (meters):")
print(delta_max)
```

```python
# Case 2: Pairwise cantilever deflection (W and L same length)
import numpy as np

# Beam constants
E = 200e9       # Young's Modulus in Pascals (Pa)
I = 0.0001      # Moment of inertia in m^4

# Arrays of loads (N) and beam lengths (m)
W = np.array([500, 1000, 1500])   # loads
L = np.array([2, 2.5, 3])         # lengths

# Pairwise maximum deflection: delta_max = (W * L^3) / (3 * E * I)
# W[i] corresponds to L[i]
delta_max = (W * L**3) / (3 * E * I)

# Display the results
print("Maximum deflection for each load-length pair (meters):")
print(delta_max)
```

```python
# Case 3: All combinations of loads and lengths using broadcasting (matrix output)
import numpy as np

# Beam constants
E = 200e9       # Young's Modulus in Pascals (Pa)
I = 0.0001      # Moment of inertia in m^4

# Arrays of loads (N) and beam lengths (m)
W = np.array([500, 1000, 1500])   # loads
L = np.array([2, 2.5, 3])         # lengths

# Reshape for broadcasting
# W as column vector, L as row vector
W_reshaped = W.reshape(-1,1)   # shape (3,1)
L_reshaped = L.reshape(1,-1)   # shape (1,3)

# Maximum deflection for all combinations
delta_max = (W_reshaped * L_reshaped**3) / (3 * E * I)

# Display the results
print("Matrix of maximum deflections (meters) for all load-length combinations:")
print(delta_max)
```

```python
# Case 4: Nested for loops with NumPy arrays (formatted, human-readable)
import numpy as np

# Beam constants
E = 200e9       # Young's Modulus in Pascals (Pa)
I = 0.0001      # Moment of inertia in m^4

# Arrays of loads (N) and beam lengths (m)
W = np.array([500, 1000, 1500])
L = np.array([2, 2.5, 3])

# Nested loops for all combinations
for w in W:
    for l in L:
        delta_max = (w * l**3) / (3 * E * I)
        print(f"Load={w} N, Length={l} m --> Max deflection={delta_max:.6f} m")
```

***matplotlib***

```python
# Case 1: Simple line plot of maximum deflection vs load for a cantilever beam
import numpy as np
import matplotlib.pyplot as plt

# Beam constants
L = 2           # Length of the beam in meters
E = 200e9       # Young's Modulus in Pascals
I = 0.0001      # Moment of inertia in m^4

# Load range
W = np.linspace(100, 2000, 50)  # Loads from 100 N to 2000 N

# Maximum deflection calculation: delta_max = (W * L^3) / (3 * E * I)
delta_max = (W * L**3) / (3 * E * I)

# Figure 1: Line plot
plt.figure(1)
plt.plot(W, delta_max, color='blue', linewidth=2)
plt.xlabel("Load W (N)")
plt.ylabel("Maximum Deflection δ_max (m)")
plt.title("Cantilever Beam Deflection vs Load (Line)")
plt.grid(True)
plt.show()
```
